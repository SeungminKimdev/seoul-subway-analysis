```python
import apikey
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from geopy.geocoders import Nominatim
from PIL import Image
import io
import json
import csv
```

```python
def getBusInfo(busId): #버스정보 얻어오기 id = bus code
    key = apikey.PBDATAKEYS['decodingKey']
    url = 'http://ws.bus.go.kr/api/rest/busRouteInfo/getStaionByRoute'
    bus_params = {'serviceKey':key,'busRouteId':busId}
    response = requests.get(url,params=bus_params)
    return response

def getLocations(address): #주소 -> 위도, 경도 변환
    geolocator = Nominatim(user_agent="location translator")
    location = geolocator.geocode(address)
    if location:
        latitude = location.latitude
        longitude = location.longitude
        return latitude, longitude
    else:
        return None
```
